4. Зробити таку транформацію можна за допомогою функції `pivot_table()`

```python
sales_pivot = pd.pivot_table(df, index='Date', columns='Branch', values='Invoice ID', aggfunc=pd.Series.nunique)
sales_pivot
```

```python
plt.figure(figsize=(10, 6))

for column in sales_pivot.columns:
    plt.plot(sales_pivot.index, sales_pivot[column], label=column)
    
plt.legend()
plt.xlabel('Date')
plt.ylabel('Quantity of sales')
plt.title('Daily sales by Branches')
```

Давайте зробимо графіки ще зручнішими для сприйняття. Хочеться їх трохи згладити. Одним з методів згладування є ковзне середнє (Moving Average - MA). Суть методу полягає в тому, що значення кожної точки дорівнює середньому значенню N попередніх точок. Це один з найпростіших методів згладжування набору статистичних даних і має кілька варіацій. [Тут](https://www.youtube.com/watch?v=7Rz_ITRIADg) непогано відео на тему. 
 
Для того, щоб порахувати ковзне середнє можемо використати функцію [pandas.DataFrame.rolling](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html).     

Используя параметры window=10 и min_periods=1 выведите сглаженные подневные графики количества продаж по филиалам. 

```python
smothed_sales_pivot=sales_pivot.rolling(window=10, min_periods=1).mean()


fig, axes = plt.subplots(3,1, figsize=(10, 10), sharex=True)

axes[0].plot(smothed_sales_pivot.index, smothed_sales_pivot['A'], label='A', color= 'green')
axes[1].plot(smothed_sales_pivot.index, smothed_sales_pivot['B'], label='B', color= 'purple')
axes[2].plot(smothed_sales_pivot.index, smothed_sales_pivot['C'], label='C', color= 'red')

axes[0].set_title('Quantity sales of A')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Quantity of sales')

axes[1].set_title('Quantity sales of B')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Quantity of sales')

axes[2].set_title('Quantity sales of C')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Quantity of sales')

fig.legend()
```

```python
filtered_df = df[(df['Unit price'] >= 90) & (df['Unit price'] <= 100)]
product_lines = filtered_df['Product line'].unique()

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)
for product_line in product_lines:
    data = filtered_df[filtered_df['Product line'] == product_line]['Unit price']
    plt.hist(data, label=product_line, **kwargs)

plt.legend()
plt.title('Distribution of Prices for Each Product Line')
plt.xlabel('Price')
plt.ylabel('Density')
```

13.2. А тепер відобразіть гістограми з однаковою кількість бінів на одному графікую Для зручності використовуйте прозорість alpha не більше 0.4 і переконайтесь, що гістограми різного кольору. Не забудьте про легенду, заголовок. 

```python
num_bins = 40
alpha = 0.4
colors = ['blue', 'green', 'red', 'orange', 'purple', 'yellow'] 

plt.figure(figsize=(10, 6))
for i, product_line in enumerate(product_lines):
    data = filtered_df[filtered_df['Product line'] == product_line]['Unit price']
    plt.hist(data, bins=num_bins, alpha=alpha, color=colors[i % len(colors)], label=product_line)
    
plt.legend()
plt.title('Distribution of Prices for Each Product Line')
plt.xlabel('Price')
plt.ylabel('Frequency')
```
